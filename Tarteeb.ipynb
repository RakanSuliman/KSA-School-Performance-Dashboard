```python
df = tarteeb_df.copy()
# Convert every missing value to NaN
text_required = [
    'city_ar','office_ar','department_ar','authority_ar',
    'region_ar','area_ar','school_name_ar',
    'exam_type_ar','exam_special_ar','gender_code','year'
]
for col in text_required:
    if col in df.columns:
        df[col] = df[col].replace(r'^\s*$', np.nan, regex=True)


```

```python
required_columns = [
    'avg','dept_avg','office_avg','area_avg','ksa_rank','std_count',
    'city_ar','office_ar','department_ar','authority_ar',
    'region_ar','area_ar','school_name_ar',
    'exam_type_ar','exam_special_ar','gender_code','year'
]
# Filtered list of the required_columns (lat/lon remain optional)
filter_req = [col for col in required_columns if col in df.columns]
tarteeb_df_cleaned = df.dropna(subset=filter_req).copy()
print("Shape before:", tarteeb_df.shape)
print("After cleaning shape:", tarteeb_df_cleaned.shape)
print("\nRemaining missing values:")
print(tarteeb_df_cleaned.isna().sum().sort_values(ascending=False).head(10))
```

```python
df = tarteeb_df_cleaned.copy()
df = df[(df['exam_type_ar'] == 'تحصيلي') & (df['exam_special_ar'] == 'علمي')]
df = df[df['gender_code'] == 'M']     # dataset is boys-only [تحصيلي,علمي]
print("Filtered (تحصيلي | علمي | M) shape:", df.shape)
```

```python
# Contextual gap (positive = underperforming vs department)
df['gap_vs_dept'] = df['dept_avg'] - df['avg']
df['gap_vs_office'] = df['office_avg'] - df['avg']
df['gap_vs_area'] = df['area_avg'] - df['avg']

# Size transforms
df['school_size_index'] = pd.cut(
    df['std_count'],
    bins=[25, 30, 200, 400, 800, 1600, 2000],
    labels=['XXS','XS','S','M','L','XL'],
    include_lowest=True
)

print("School Sizes Index distribution:")
print(df['school_size_index'].value_counts(dropna=False))
```

```python
#  Save the new cleaned data
cleaned_file_path_v2 = "tarteeb_cleaned_v2.csv"
tarteeb_df_cleaned.to_csv(cleaned_file_path_v2, index=False, encoding='utf-8-sig')

print(f"\nCleaned data (ignoring lat/lon for dropping) saved to: {cleaned_file_path_v2}")
print("\n--- Missing Values Count in New Dataframe (Notice lat/lon still have them) ---")
print(tarteeb_df_cleaned.isnull().sum().sort_values(ascending=False))
```

```python
# Filtering utility
def pick(year=None, exam=None, special=None, gender=None, base=None):
    _df = df if base is None else base
    m = pd.Series(True, index=_df.index)

    if year is not None and 'year' in _df:
        m &= _df['year'].astype(str) == str(year)

    if exam is not None and 'exam_type_ar' in _df:
        m &= _df['exam_type_ar'] == exam        # "تحصيلي" | "قدرات"

    if special is not None and 'exam_special_ar' in _df:
        m &= _df['exam_special_ar'] == special  # "علمي" | "أدبي"

    if gender is not None and 'gender_code' in _df:
        m &= _df['gender_code'] == gender       # "M" | "F"

    return _df[m].copy()
```

```python
df_geo = tarteeb_df_cleaned.copy()
df_geo = df.dropna(subset=['lat', 'lon'])
print(f"Plotting {df_geo.shape[0]} schools that have lat/lon data.")
```

```python
tarteeb_df['avg'] = pd.to_numeric(tarteeb_df['avg'], errors='coerce')

gender_labels = {'M': 'Male', 'F': 'Female'}

g = tarteeb_df.groupby('gender_code')['avg']
means = g.mean().rename(index=gender_labels)
ses   = g.sem().rename(index=gender_labels)

x = np.arange(len(means))

plt.figure()
plt.bar(x, means.values, yerr=ses.values, capsize=4, edgecolor='black')
plt.xticks(x, means.index)
plt.ylabel('Average grade')
plt.title('Average grade by gender')
plt.tight_layout()
plt.show()
```
